## **Merge Two Files**

```python
from PyPDF2 import PdfReader, PdfWriter

file1 = PdfReader('file1.pdf')
file2 = PdfReader('file2.pdf')

output = PdfWriter()

output.append(file1)
output.append(file2)

output.write('merged_file.pdf')
```

## **Extract and Save Required Pages from PDF**

```python
from PyPDF2 import PdfReader, PdfWriter

inputpdf = PdfReader(open("merged_file.pdf", "rb"))
total_pages = len(inputpdf.pages)
print('Total pages are ', total_pages)

is_user_input_invalid = False

user_input = input("Enter page numbers separated by comma (example: 1, 2, 3): ")
page_numbers = user_input.split(', ')

# Convert page numbers to integers and check validity
pages = []
for x in page_numbers:
  page_number = int(x)
  if page_number <= total_pages:
      pages.append(page_number)
  else:
      is_user_input_invalid = True

if is_user_input_invalid:
  print("Some page numbers are invalid or out of range.")
else:
  output = PdfWriter()
  for page in pages:
    output.add_page(inputpdf.pages[page])
  output.write('new_file.pdf')
```
